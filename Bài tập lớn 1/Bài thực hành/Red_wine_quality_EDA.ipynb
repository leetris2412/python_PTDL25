### 4. Khám phá dữ liệu

```python
# Get the mean of the data
data_mean = np.mean(df["alcohol"])
print("Mean of alcohol:", data_mean)
```

```python
# Get the median of the data
data_median = np.median(df["alcohol"])
print("Median of alcohol:", data_median)
```

```python
# Get the mode of the data
data_mode = stats.mode(df["alcohol"])
print("Mode of alcohol:", data_mode)
```

```python
# Obtain the variance of the data
data_variance = np.var(df["alcohol"])
print("Variance of alcohol:", data_variance)
```

```python
# Obtain the standard deviation of the data
data_sd = np.std(df["alcohol"])
print("Standard Deviation of alcohol:", data_sd)
```

```python
# Compute the maximum and minimum values of the data
data_max = np.max(df["alcohol"])
data_min = np.min(df["alcohol"])
print("Maximum value of alcohol:", data_max)
print("Minimum value of alcohol:", data_min)
```

```python
# Obtain the 60th percentile of the data
data_percentile = np.percentile(df["alcohol"],60)
print("60th Percentile of alcohol:", data_percentile)
```

```python
# Obtain the quartiles of the data
data_quartile = np.quantile(df["alcohol"],0.75)
print("75th Percentile (Q3) of alcohol:", data_quartile)
```

```python
# Get the IQR of the data
data_IQR = stats.iqr(df["alcohol"])
print("Interquartile Range (IQR) of alcohol:", data_IQR)
```
