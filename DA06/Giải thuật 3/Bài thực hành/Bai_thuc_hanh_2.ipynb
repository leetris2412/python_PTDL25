# Xây dựng mô hình Naïve ngây thơ trên tập dữ liệu mushroom.

```python
# Import các thư viện cần thiết
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler 

# THAY ĐỔI QUAN TRỌNG: Dùng GaussianNB cho dữ liệu số, thay vì MultinomialNB
from sklearn.naive_bayes import MultinomialNB
```

```python
df = pd.read_csv('datasets/mushrooms.csv')
```

```python
df.head()
```

#### Xử lý dữ liệu

```python
# Tách Features (X) và Target (y)
# y (target) là cột 'class' (e = edible, p = poisonous)
y = df['class']
# X (features) là tất cả các cột còn lại
X = df.drop(columns=['class'])
```

```python
# Chia dữ liệu thành tập train và test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,      # 20% test
    random_state=42     # Giữ nguyên random_state cho nhất quán
)
```

#### Vector hóa

```python
# 1. Mã hóa X_train
# Dùng pd.get_dummies() để biến tất cả cột chữ trong X_train thành số (0/1)
X_train_vectorized = pd.get_dummies(X_train)

# 2. Mã hóa X_test
# Dùng pd.get_dummies() cho X_test
X_test_vectorized = pd.get_dummies(X_test)

# 3. Đảm bảo X_test có cùng các cột với X_train
# (Có thể 1 giá trị nào đó chỉ xuất hiện ở train mà không ở test, hoặc ngược lại)
X_train_cols, X_test_cols = X_train_vectorized.align(X_test_vectorized, join='left', axis=1, fill_value=0)
```

#### Xây dựng mô hình Naïve Bayes

```python
# 1. Khởi tạo mô hình
classifier = MultinomialNB() # 

# 2. Huấn luyện (fit) mô hình
classifier.fit(X_train_cols, y_train)
```

#### Đánh giá mô hình

```python
# 1. Dự đoán trên tập test đã mã hóa
y_pred = classifier.predict(X_test_cols)

# 2. Đánh giá mô hình
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
```

```python
# 3. In kết quả
print(f"\nAccuracy: {accuracy:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(classification_rep)
```
